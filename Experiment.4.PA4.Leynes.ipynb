```python
#Import the required related library
import pandas as pd
import matplotlib.pyplot as plt
```

## ECE Board Exam Data üíΩ

```python
#Load the board2.xlsx, csv file into a data frame using pandas.
enggeceexamds = pd.read_excel('board2.xlsx')
enggeceexamds
```

## Problem 1Ô∏è‚É£ Data Frames üêº

#### **1Ô∏èa.** Create the following data frames based on the format provided: 
#### Filename: `Instru [‚ÄúName‚Äù, ‚ÄúGEAS‚Äù, ‚ÄúElectronics >70‚Äù]; where track is constant as Instrumentation and hometown Luzon`

```python
#Data filtering using pandas with conditions
Instru = enggeceexamds.loc[(enggeceexamds ['Track'] == 'Instrumentation') & 
                           (enggeceexamds ['Hometown'] == 'Luzon') &
                           (enggeceexamds ['Electronics'] > 70), ['Name', 'GEAS', 'Electronics']]
Instru
```

#### **1Ô∏èb.** Create the following data frames based on the format provided: 
#### Filename: `Mindy = [ ‚ÄúName‚Äù, ‚ÄúTrack‚Äù, ‚ÄúElectronics‚Äù, ‚ÄúAverage >=55‚Äù]; where hometown is constant as Mindanao and gender Female`

```python
#Calculates the average for all subjects
enggeceexamds['Average'] = (enggeceexamds['Math'] + enggeceexamds['Electronics'] + 
                            enggeceexamds['GEAS'] + enggeceexamds['Communication']) / 4

#Data filtering using pandas with conditions
Mindy = enggeceexamds.loc[(enggeceexamds['Hometown']=='Mindanao') & 
                          (enggeceexamds['Gender']=='Female') & 
                          (enggeceexamds['Average']>=55), ('Name','Track','Electronics','Average')]
Mindy
```

## Problem 2Ô∏è‚É£ Visualization üìä

#### Create a visualization that shows how the different features contributes to average grade. 
#### `Does chosen track in college, gender, or hometown contributes to a higher average score?`

```python
#Transform the data to a graph (Track = Blue, Gender = Orange, Hometown = Green)
plt.figure(figsize=(15, 5))
plt.bar(enggeceexamds['Track'], enggeceexamds['Average'])
plt.bar(enggeceexamds['Gender'], enggeceexamds['Average'])
plt.bar(enggeceexamds['Hometown'], enggeceexamds['Average'])

#Add titles and labels
plt.title('Average Grades by Track, Gender, and Hometown')
plt.xlabel('Categories')
plt.ylabel('Average Grade')
```
